```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
#un terzo ogni, un terzo accomulo 8 poi 12 diverge
#bs 32 48 64
#e 10 15 20
#corrected
df = pd.read_csv('Project/data_s18_2.csv')
df_e = pd.read_csv('Project/data_s18_2_e.csv')
df_eie = pd.read_csv('Project/data_eachIter_e.csv')
```

```python
df
```

```python
n20_df = df.query('num_epochs == 20')
seaborn.set_style("whitegrid")
g2 = seaborn.relplot(data=n20_df, x='num_nodes', y='acc', hue='bs', kind="line", legend = True, err_style="bars")
g2.fig.suptitle("20 epochs", y = 0.95)
g2.set(ylabel= "Accuracy", xlabel="num_nodes")
plt.legend(labels=["Batch size = 32", "Batch size = 48", "Batch size = 64"])
plt.show()
```

```python
n20_df = df.query('num_epochs == 20')
seaborn.set_style("whitegrid")
g1 = seaborn.relplot(data=n20_df, x='num_nodes', y='time', hue='bs', kind="line", legend = False, err_style="bars")
g1.fig.suptitle("20 epochs", y = 0.95)
g1.set(ylabel= "Time(s)", xlabel="num_nodes")
plt.legend(labels=[ "Batch size = 50", "Batch size = 64"])
plt.show()
```

```python
n2_df = df_e.query('num_nodes == 2 or num_nodes == 4  or num_nodes ==6  or num_nodes ==12 or num_nodes ==14')
seaborn.set_style("whitegrid")
g2 = seaborn.relplot(data=n2_df, x='num_epochs', y='acc', hue='num_nodes', kind="line", legend = False, err_style="bars")
g2.fig.suptitle("Batch size = 64", y = 0.95)
g2.set(ylabel= "Accuracy", xlabel="Epochs")
plt.legend(labels=["2 nodes", "4 nodes",  "6 nodes","12 nodes","14 nodes"])
plt.savefig('s18_e.png')
plt.show()

ei_df = df_eie.query('num_nodes == 2 or num_nodes == 4  or num_nodes ==6  or num_nodes ==12 or num_nodes ==14')
seaborn.set_style("whitegrid")
g2 = seaborn.relplot(data=ei_df, x='num_epochs', y='acc', hue='num_nodes', kind="line", legend = True, err_style="bars")
g2.fig.suptitle("Batch size = 64", y = 0.95)
g2.set(ylabel= "Accuracy", xlabel="Epochs")
#plt.legend(labels=["Batch size = 32", "Batch size = 48", "Batch size = 64"])
plt.show()
```
