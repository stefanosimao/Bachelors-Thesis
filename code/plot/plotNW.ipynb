```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
```

```python
df.plot(kind='scatter',x='num_workers',y='time',color='red')
plt.show()
```

```python
n2_df = df.query('num_nodes == 2')
seaborn.set_style("whitegrid")
g2 = seaborn.relplot(data=n2_df, x='num_workers', y='time', hue='bs', kind="line", legend = False)
g2.fig.suptitle("2 nodes", y = 0.95)
g2.set(ylabel= "Time(s)", xlabel="num_workers")
plt.legend(labels=["Batch size = 50","Batch size = 150", "Batch size = 250"])
plt.savefig('n2_nw.png')
plt.show()
```

```python
num_workers_cases = {
      2: {50: 3, 150: 8, 250: 10},
      4: {50: 3, 150: 8, 250: 9},
      6: {50: 3, 150: 6, 250: 7},
      8: {50: 3, 150: 6, 250: 5},
      10: {50: 3, 150: 5, 250: 5},
      12: {50: 3, 150: 4, 250: 4},
      14: {50: 3, 150: 4, 250: 4},
      16: {50: 3, 150: 4, 250: 3},
}
def adapt_num_workers(world_size, batch_size):
      return num_workers_cases.get(world_size, {}).get(batch_size, 6)

n_w = adapt_num_workers(4, 250)
print(n_w)
```
