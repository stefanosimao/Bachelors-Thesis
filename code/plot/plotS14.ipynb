```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
#un terzo ogni, due terzi accomulo 4
#bs 48 64 80
#e 20 25
#corrected
df = pd.read_csv('Project/data_s14.csv')
```

```python
df
```

```python
n20_df = df.query('num_epochs == 20')
seaborn.set_style("whitegrid")
g1 = seaborn.relplot(data=n20_df, x='num_nodes', y='acc', hue='bs', kind="line", legend = True, err_style="bars")
g1.fig.suptitle("20 epochs", y = 0.95)
g1.set(ylabel= "Accuracy", xlabel="num_nodes")
plt.legend(labels=["Batch size = 48", "Batch size = 64", "Batch size = 80"])
plt.show()

n25_df = df.query('num_epochs == 25')
seaborn.set_style("whitegrid")
g2 = seaborn.relplot(data=n25_df, x='num_nodes', y='acc', hue='bs', kind="line", legend = False, err_style="bars")
g2.fig.suptitle("25 epochs", y = 0.95)
g2.set(ylabel= "Accuracy", xlabel="num_nodes")
plt.legend(labels=["Batch size = 48", "Batch size = 64", "Batch size = 80"])
plt.show()
```

```python
n20_df = df.query('num_epochs == 20')
seaborn.set_style("whitegrid")
g1 = seaborn.relplot(data=n20_df, x='num_nodes', y='time', hue='bs', kind="line", legend = False, err_style="bars")
g1.fig.suptitle("20 epochs", y = 0.95)
g1.set(ylabel= "Time(s)", xlabel="num_nodes")
plt.legend(labels=["Batch size = 48", "Batch size = 64", "Batch size = 80"])
plt.show()

n25_df = df.query('num_epochs == 25')
seaborn.set_style("whitegrid")
g2 = seaborn.relplot(data=n25_df, x='num_nodes', y='time', hue='bs', kind="line", legend = False, err_style="bars")
g2.fig.suptitle("25 epochs", y = 0.95)
g2.set(ylabel= "Time(s)", xlabel="num_nodes")
plt.legend(labels=["Batch size = 48", "Batch size = 64", "Batch size = 80"])
plt.show()
```
