```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
#un terzo ogni, un terzo accomulo 4 poi 8 diverge
df = pd.read_csv('Project/data_s9.csv')
```

```python
df
```

```python
n20_df = df.query('num_epochs == 20')
seaborn.set_style("whitegrid")
g1 = seaborn.relplot(data=n20_df, x='num_nodes', y='acc', hue='bs', kind="line", legend = False, err_style="bars")
g1.fig.suptitle("20 epochs", y = 0.95)
g1.set(ylabel= "Accuracy", xlabel="num_nodes")
plt.legend(labels=["Batch size = 50","Batch size = 150", "Batch size = 250"])
plt.show()

n40_df = df.query('num_epochs == 40')
seaborn.set_style("whitegrid")
g2 = seaborn.relplot(data=n40_df, x='num_nodes', y='acc', hue='bs', kind="line", legend = False, err_style="bars")
g2.fig.suptitle("40 epochs", y = 0.95)
g2.set(ylabel= "Accuracy", xlabel="num_nodes")
plt.legend(labels=["Batch size = 50","Batch size = 150", "Batch size = 250"])
plt.show()

n60_df = df.query('num_epochs == 60')
seaborn.set_style("whitegrid")
g3 = seaborn.relplot(data=n60_df, x='num_nodes', y='acc', hue='bs', kind="line", legend = False, err_style="bars")
g3.fig.suptitle("60 epochs", y = 0.95)
g3.set(ylabel= "Accuracy", xlabel="num_nodes")
plt.legend(labels=["Batch size = 50","Batch size = 150", "Batch size = 250"])
plt.show()
```

```python
n20_df = df.query('num_epochs == 20')
seaborn.set_style("whitegrid")
g1 = seaborn.relplot(data=n20_df, x='num_nodes', y='time', hue='bs', kind="line", legend = False, err_style="bars")
g1.fig.suptitle("20 epochs", y = 0.95)
g1.set(ylabel= "Time(s)", xlabel="num_nodes")
plt.legend(labels=["Batch size = 50","Batch size = 150", "Batch size = 250"])
plt.show()

n40_df = df.query('num_epochs == 40')
seaborn.set_style("whitegrid")
g2 = seaborn.relplot(data=n40_df, x='num_nodes', y='time', hue='bs', kind="line", legend = False, err_style="bars")
g2.fig.suptitle("40 epochs", y = 0.95)
g2.set(ylabel= "Time(s)", xlabel="num_nodes")
plt.legend(labels=["Batch size = 50","Batch size = 150", "Batch size = 250"])
plt.show()

n60_df = df.query('num_epochs == 60')
seaborn.set_style("whitegrid")
g3 = seaborn.relplot(data=n60_df, x='num_nodes', y='time', hue='bs', kind="line", legend = False, err_style="bars")
g3.fig.suptitle("60 epochs", y = 0.95)
g3.set(ylabel= "Time(s)", xlabel="num_nodes")
plt.legend(labels=["Batch size = 50","Batch size = 150", "Batch size = 250"])
plt.show()
```
