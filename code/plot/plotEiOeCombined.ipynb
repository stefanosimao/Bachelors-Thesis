```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
```

```python
df_oe = pd.read_csv('Project/data_onlyEnd.csv')
df_ei = pd.read_csv('Project/data_eachIter.csv')
df_s = pd.read_csv('Project/data_s18.csv')
```

```python
df_oe['strategy'] = 1
df_ei['strategy'] = 2
df_s['strategy'] = 3
```

```python
e20_50_df_oe = df_oe.query('num_epochs == 20 & bs == 50')
e20_50_df_ei = df_ei.query('num_epochs == 20 & bs == 50')
e20_64_df_s = df_s.query('num_epochs == 20 & bs == 64')
comb1 = [e20_50_df_oe, e20_50_df_ei, e20_64_df_s]
comb20_50 = pd.concat(comb1,ignore_index=True)
seaborn.set_style("whitegrid")
g1 = seaborn.relplot(data=comb20_50, x='num_nodes', y='acc', hue='strategy', kind="line", legend = False, err_style="bars")
g1.fig.suptitle("20 epochs", y = 0.96)
g1.set(ylabel= "Accuracy", xlabel="num_nodes")
plt.legend(labels=["OnlyEnd","EachIter","8,12"])
plt.savefig('confr.png')
plt.show()

e20_150_df_oe = df_oe.query('num_epochs == 20 & bs == 150')
e20_150_df_ei = df_ei.query('num_epochs == 20 & bs == 150')
comb2 = [e20_150_df_oe, e20_150_df_ei]
comb20_150 = pd.concat(comb2,ignore_index=True)
seaborn.set_style("whitegrid")
g2 = seaborn.relplot(data=comb20_150, x='num_nodes', y='acc', hue='strategy', kind="line", legend = False, err_style="bars")
g2.fig.suptitle("20 epochs & 150 batch size", y = 0.96)
g2.set(ylabel= "Accuracy", xlabel="num_nodes")
plt.legend(labels=["OnlyEnd","EachIter"])
plt.show()

e20_250_df_oe = df_oe.query('num_epochs == 20 & bs == 250')
e20_250_df_ei = df_ei.query('num_epochs == 20 & bs == 250')
comb3 = [e20_250_df_oe, e20_250_df_ei]
comb20_250 = pd.concat(comb3,ignore_index=True)
seaborn.set_style("whitegrid")
g3 = seaborn.relplot(data=comb20_250, x='num_nodes', y='acc', hue='strategy', kind="line", legend = False, err_style="bars")
g3.fig.suptitle("20 epochs & 250 batch size", y = 0.96)
g3.set(ylabel= "Accuracy", xlabel="num_nodes")
plt.legend(labels=["OnlyEnd","EachIter"])
plt.show()
```
